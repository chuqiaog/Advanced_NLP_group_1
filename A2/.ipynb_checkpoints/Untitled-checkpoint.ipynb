```python
trainfile = 'data/en_ewt-up-train.conllu'
devfile = 'data/en_ewt-up-dev.conllu'
testfile = 'data/en_ewt-up-test.conllu'
```

```python
trainlist = read_conll(trainfile)
devlist = read_conll(devfile)
testlist = read_conll(testfile)
```

```python
preprocessed_train = preprocess_list(trainlist)
preprocessed_dev = preprocess_list(devlist)
preprocessed_test = preprocess_list(testlist)
```

```python
def get_mappings_dict(preprocessed_train,preprocessed_dev,preprocessed_test):
    preps = preprocessed_train+preprocessed_dev+preprocessed_test
    arg_list=[]
    for sent in preps:
        for token in sent:
            arg_list.append(token['ARG'])
    arg_list = set(sorted(arg_list))
    map_dict = dict(zip(set(arg_list), range(len(set(arg_list)))))
    map_dict_reverse = {v: k for k, v in map_dict.items()}
    return map_dict, map_dict_reverse
    
label_dict, label_dict_rev = get_mappings_dict(preprocessed_train,preprocessed_dev,preprocessed_test)
```

```python
def create_word_sentlist(pre_list):
    '''
    '''
    word_sentlist = []
    for sentence in pre_list:
        featdict = {}
        wordlist = []
        #sent = ''
        #is_arg = []
        args = []
        pred = []
        for token in sentence:
            wordlist.append(token['form'])
            #sent = sent + ' '+token['form']
            if token['V'] == 'V':
                pred.append(token['form'])
            args.append(label_dict[token['ARG']])
            #if token['ARG'] == '_':
            #    is_arg.append(0)
            #else: is_arg.append(1)
        #if pred != '':
            #wordlist.append('[SEP]')
            #wordlist.append(pred)
            #args.append(label_dict['_'])
            #args.append(label_dict['_'])
        featdict['tokens'],featdict['srl_arg_tags'],featdict['pred'] = wordlist,args,pred
        #featdict['sentence'],featdict['srl_arg_tags'],featdict['pred'] = sent,is_arg,pred
        word_sentlist.append(featdict)
    
    return word_sentlist
```

```python
trainsent = create_word_sentlist(preprocessed_train)
devsent = create_word_sentlist(preprocessed_dev)
testsent = create_word_sentlist(preprocessed_test)
```

```python
import datasets
```

```python
trainds = datasets.Dataset.from_list(trainsent)
devds = datasets.Dataset.from_list(devsent)
testds = datasets.Dataset.from_list(testsent)
```

```python
testds[0]
```

```python
task = "srl"
model_checkpoint = "bert-base-uncased" # bert-base-uncased for better percision, distilbert-base-uncased for faster run
batch_size = 16
import transformers
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
assert isinstance(tokenizer, transformers.PreTrainedTokenizerFast)
```

```python
def tokenize_and_align_labels(examples, label_all_tokens=True):
    
    tokenized_inputs = tokenizer(examples["tokens"],examples['pred'], truncation=True, is_split_into_words=True)

    labels = []
    all_word_ids = []
    for i, label in enumerate(examples[f"{task}_arg_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            # Special tokens have a word id that is None. We set the label to -100 so they are automatically
            # ignored in the loss function.
            if (word_idx is None):
                label_ids.append(-100)
            # We set the label for the first token of each word.
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            # For the other tokens in a word, we set the label to either the current label or -100, depending on
            # the label_all_tokens flag.
            else:
                label_ids.append(label[word_idx] if label_all_tokens else -100)
            previous_word_idx = word_idx

        labels.append(label_ids)
        all_word_ids.append(word_ids)
    
    
    tokenized_inputs['word_ids'] = all_word_ids
    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
tokenized_train = trainds.map(tokenize_and_align_labels, batched=True)
tokenized_dev = devds.map(tokenize_and_align_labels, batched=True)
tokenized_test = testds.map(tokenize_and_align_labels, batched=True)
```

```python
tokenized_test[2]
```

```python
from transformers import AutoModelForTokenClassification, TrainingArguments, Trainer
model = AutoModelForTokenClassification.from_pretrained(model_checkpoint, num_labels=len(label_dict))
model_name = model_checkpoint.split("/")[-1]
args = TrainingArguments(
    f"{model_name}-finetuned-{task}",
    evaluation_strategy = "epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=5,
    weight_decay=0.01,
    push_to_hub=False,
)
from transformers import DataCollatorForTokenClassification
data_collator = DataCollatorForTokenClassification(tokenizer)
label_list = sorted(label_dict_rev)
from datasets import load_metric
metric = load_metric("seqeval")
```

```python
import numpy as np
def remove_ignored_index(predictions,labels):
    actual_predictions = [
        [label_list[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    actual_labels = [
        [label_list[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    return actual_predictions, actual_labels
    
def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2) # Most possible label

    # Remove ignored index (special tokens)
    actual_predictions, actual_labels = remove_ignored_index(predictions,labels)
    
    results = metric.compute(predictions=actual_predictions, references=actual_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
trainer = Trainer(
    model,
    args,
    train_dataset=tokenized_dev,
    eval_dataset=tokenized_test,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```
